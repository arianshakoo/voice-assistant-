```python
import re
from hazm import *
import pandas as pd
import numpy as np
```

```python
df=pd.read_json('https://raw.githubusercontent.com/sajjjadayobi/PersianQA/main/dataset/pqa_train.json')
```

```python
def clean_text(text):
  if text!='':
      text = re.sub(r'^http?:\/\/*[\r\n]*', '', text, flags=re.MULTILINE)
      text = re.sub(r'^https?:\/\/*[\r\n]*', '', text, flags=re.MULTILINE)
      text=re.sub("[a-zA-Z0-9]", '' , text)
      pattern = r'/'
      text = re.sub(pattern, '', text )
      pattern = r'#'
      text = re.sub(pattern, '', text )
      pattern = r'@'
      text = re.sub(pattern, '', text )
      pattern = r'_'
      text = re.sub(pattern, ' ', text )
      text=re.sub(r'pic.twitter.com[\w]*',"", text)
      pattern = r':'
      text = re.sub(pattern, ' ', text )
      return text
```

```python
x=[]
y=[]
for i in range(901):
  for j in range(len(df.iloc[i]['data']['paragraphs'][0]['qas'])):
    x.append(df.iloc[i]['data']['paragraphs'][0]['qas'][j]['question'])
    if df.iloc[i]['data']['paragraphs'][0]['qas'][j]['answers']!=[]:
      text=df.iloc[i]['data']['paragraphs'][0]['qas'][j]['answers'][0]['text']
    y.append(clean_text(text))
```

```python
numb = 0
for i in y:
  numb+=1
  name = str(numb)+"_ans"+".txt"
  f = open(name,"w+")
  f.write(i)
  f.close()
```
